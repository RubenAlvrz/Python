```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel('mpg.xlsx')

df.head()
```

```python
plt.figure(figsize= (10,5))

sns.scatterplot(x= df['mpg'], y= df['horsepower'] , data = df, hue = 'origin' )            

plt.title("Horsepower Vs MPG",     
          y = 1.05,                  
          size = 15);   
```

```python
plt.figure(figsize= (15,8))

sns.scatterplot(x= df['mpg'], y= df['weight'] , data = df, hue = 'origin', style = 'cylinders')            

plt.title("Weight Vs MPG",     
          y = 1.05,                  
          size = 15);  
```
