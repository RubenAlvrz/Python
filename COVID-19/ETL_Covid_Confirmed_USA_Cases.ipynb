```python
#Import the pandas library and read csv file
import pandas as pd

df = pd.read_csv('covid_confirmed_usafacts.csv')

df.head()
```

```python
# generate a descriptive statistics for the dataframe
df.describe()
```

```python
#From analyzing the dataframe we found that some countyFIPS are equal to 0.
# We will go ahead and select the countyFIPS that are not equal to 0

df1 = df[df['countyFIPS'] != 0]

df1.head()
```

```python
#Created a seperate table with dates and the number of cases
df2 = df1.iloc[:,4:]

df2.tail(10)
```

```python
# From df2 we will use "diff" this calculates the difference of a Dataframe element compared with another element in the Dataframe 
# (default is element in previous row)

df3 = df2.diff(axis = 1)

df3.head()
```

```python
# Then create a new table with countyFIPS,County Name,State,StateFIPS columns

df_info = df.iloc[:,:4]

df_info.head()
```

```python
#Then we will merge df_info dataset back to dataset df3

df_merge = df_info.merge(df3,left_index=True, right_index=True)

df_merge.head()
```

```python
#Next we will unpivot the table by using pandas melt

df_melt = pd.melt(frame=df_merge ,id_vars = ['countyFIPS','County Name','State','StateFIPS'], var_name = "Date")

df_melt.head()
```

```python
# View df shape
df_melt.shape
```

```python
# fill in the null values and covert to int

df_melt['value'] = df_melt['value'].fillna('0').astype(int)
```

```python
#Export to csv file
df_melt.to_csv('etl_covid_confirmed_usafacts.csv')
```
